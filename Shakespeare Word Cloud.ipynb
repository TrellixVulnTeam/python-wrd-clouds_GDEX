```python
import pathlib
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
path = pathlib.Path.cwd()  /'hamlet.txt'
print(path)
```

```python
text2 = path.read_text()
text1 = path.read_text()
wordcloud = WordCloud().generate(text2)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('./hamlet.png', dpi=600, facecolor='w', format='png')
plt.show()

```

```python
mask = np.array(Image.open('bill-spears-2.jpg'))
stopwords = set(STOPWORDS)
plt.imshow(mask)
```

```python
wordcloud_2 = WordCloud(stopwords=stopwords, background_color='white', mode="RGBA", max_words=10000).generate(text1)
```
